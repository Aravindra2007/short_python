# formal and Actual Parameters

## formal Parameters  
    -> the variable names listed inside the def statement they act as placeholders
## Actual Parameters 
    -> the real values passed when calling the function

```python
def add(a,b):   ## formal parameters(called Method)
    return a+b

add(10,20)   ## actual parameters (calling method)
```
