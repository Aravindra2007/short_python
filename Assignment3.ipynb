2. A company stores employee records in a hierarchical structure where each employee
    reports to exactly one manager except the CEO. The management wants to determine the
    total number of employees under a particular manager, including employees reporting
    indirectly through multiple levels.
    Develop a Python program to solve this problem using recursion. Explain how recursive
    calls are executed and how the recursion terminates.

```python
def office(manager,hierarchy):

    count = 0
    for employe in hierarchy.get(manager,[]):
        count += 1

        count += office(employe,hierarchy)
    return count

hierarchy = {"CEO":["manager1","manager2","manager3","manager4","manager5"],
            "manager1":["emp1","emp2","emp3"],
             "manager2":["emp4","emp5"],
             "manager3":["emp6","emp8"],
             "manager4":["emp7","emp11","emp9"],
             "manager5":["emp10"],
             "emp1":["emp12","emp13"],
             "emp5":["emp16","emp14"],
             "emp10":["emp15"]
             }

emp = input("enter the emp name : ")
total = office(emp,hierarchy)
print(f"Total {total} members working under {emp}")
```

```python
file = open("marks.txt", "r")

marks = []
for mark in file.read().split():
    marks.append(int(mark))

file.close()

total = sum(marks)
length = len(marks)
highest = max(marks)
lowest = min(marks)
avg = total/length

print("Total Students:", length)
print("Highest Mark:", highest)
print("Lowest Mark:", lowest)
print("Average Mark:", avg)
```
